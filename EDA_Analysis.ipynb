# Sales

```python
# load exchange rates table
path =r"D:\Datascience\Project\Global_electronics\Sales.csv"
df_sales= pd.read_csv(path,encoding = "latin-1")
df_sales
```

```python
# check data types 
df_sales.info()
```

```python
# Assuming df_sales is your sales DataFrame

# Step 1: Convert to datetime format
df_sales['Order Date'] = pd.to_datetime(df_sales['Order Date'], errors='coerce')
df_sales['Delivery Date'] = pd.to_datetime(df_sales['Delivery Date'], errors='coerce')

# Step 2: Calculate Delivery Time (temporary timedelta)
df_sales['Delivery Time'] = df_sales['Delivery Date'] - df_sales['Order Date']

# Step 3: Calculate average delivery time (excluding nulls)
avg_delivery_days = df_sales['Delivery Time'].mean()

# Step 4: Fill missing Delivery Dates using average
df_sales['Delivery Date'] = df_sales['Delivery Date'].fillna(
    df_sales['Order Date'] + avg_delivery_days
)

# Step 5: Recalculate Delivery Time as integer (in days)
df_sales['Delivery Time'] = (df_sales['Delivery Date'] - df_sales['Order Date']).dt.days

# Step 6: Convert date columns to date only (no time)
df_sales['Order Date'] = df_sales['Order Date'].dt.date
df_sales['Delivery Date'] = df_sales['Delivery Date'].dt.date
```

```python
# cheking missinng values 
df_sales.isnull().sum()
```

```python
df_sales.to_csv('cleaned_sales.csv', index=False)
df_sales
```

```python

# Load all cleaned CSVs
df_customers = pd.read_csv('cleaned_customers.csv')
df_stores = pd.read_csv('cleaned_stores.csv')
df_sales = pd.read_csv('cleaned_sales.csv')
df_products = pd.read_csv('cleaned_products.csv')
df_exchange = pd.read_csv('cleaned_exchange.csv')

# Rename customer location columns to avoid conflicts
df_customers.rename(columns={
    'State': 'Customer State',
    'Country': 'Customer Country'
}, inplace=True)

# Rename store location columns to avoid conflicts
df_stores.rename(columns={
    'State': 'Store State',
    'Country': 'Store Country'
}, inplace=True)

# Merge sales with customers (on CustomerKey)
merged_df = df_sales.merge(df_customers, on='CustomerKey', how='left')

# Merge with products (on ProductKey)
merged_df = merged_df.merge(df_products, on='ProductKey', how='left')

# Merge with stores (on StoreKey)
merged_df = merged_df.merge(df_stores, on='StoreKey', how='left')

# Convert dates to datetime
df_exchange['Date'] = pd.to_datetime(df_exchange['Date'])
merged_df['Order Date'] = pd.to_datetime(merged_df['Order Date'])

# Merge exchange rate data
merged_df = merged_df.merge(
    df_exchange,
    left_on=['Currency Code', 'Order Date'],
    right_on=['Currency Code', 'Date'],
    how='left'
)

# Drop redundant exchange rate date column
merged_df.drop(columns=['Date'], inplace=True)


print("Merging completed and saved as 'merged_global_electronics.csv'")
```

```python
# moving most relevant columns to front 
cols = ['Order Number', 'Order Date', 'CustomerKey', 'Customer Country', 'StoreKey', 'Store Country', 
        'ProductKey', 'Product Name', 'Category', 'Quantity', 'Currency Code', 'Exchange', 
        'Unit Price USD', 'Unit Cost USD', 'Profit per Unit', 'Profit Margin (%)']

# Keep order for export (does not drop any columns)
merged_df = merged_df[[col for col in cols if col in merged_df.columns] + 
                      [col for col in merged_df.columns if col not in cols]]
```

```python
# Save merged file
merged_df.to_csv('merged_global_electronics.csv', index=False)
```

# Create an sql database

```python
import mysql.connector
from mysql.connector import Error

try:
    connection = mysql.connector.connect(
        host='localhost',
        user='root',
        password='root',
        database='Datasparkdb'
    )

    if connection.is_connected():
        print("Connected to Datasparkdb")
        cursor = connection.cursor()

        # Create the reordered and renamed table
        cursor.execute("""
            CREATE TABLE merged_global_electronics (
                `Order Number` INT,
                `Order Date` DATE,
                `CustomerKey` INT,
                `Customer Country` VARCHAR(100),
                `StoreKey` INT,
                `Store Country` VARCHAR(100),
                `ProductKey` INT,
                `Product Name` VARCHAR(255),
                `Category` VARCHAR(100),
                `Quantity` INT,
                `Currency Code` VARCHAR(255),
                `Exchange` FLOAT,
                `Unit Price USD` FLOAT,
                `Unit Cost USD` FLOAT,
                `Profit per Unit` FLOAT,
                `Profit Margin (%)` FLOAT,
                `Line Item` INT,
                `Delivery Date` DATE,
                `Delivery Time` INT,
                `Gender` VARCHAR(10),
                `Name` VARCHAR(255),
                `City` VARCHAR(100),
                `Customer State` VARCHAR(100),
                `Continent` VARCHAR(50),
                `Birthday` DATE,
                `Age` INT,
                `Brand` VARCHAR(100),
                `Color` VARCHAR(50),
                `Subcategory` VARCHAR(100),
                `CategoryKey` INT,
                `Store State` VARCHAR(100),
                `Square Meters` FLOAT,
                `Open Date` DATE
            );
        """)
        print("'merged_global_electronics' table created successfully with final structure.")

except Error as e:
    print(f" Error: {e}")
```

```python


# Load the merged CSV
df = pd.read_csv("merged_global_electronics.csv")

# Convert date columns to datetime
date_cols = ['Order Date', 'Delivery Date', 'Birthday', 'Open Date']
for col in date_cols:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors='coerce')

# Rename and reorder columns
df.rename(columns={
    'State_x': 'Customer State',
    'Country_x': 'Customer Country',
    'State_y': 'Store State',
    'Country_y': 'Store Country'
}, inplace=True)

column_order = [
    'Order Number', 'Order Date', 'CustomerKey', 'Customer Country',
    'StoreKey', 'Store Country', 'ProductKey', 'Product Name', 'Category',
    'Quantity', 'Currency Code', 'Exchange', 'Unit Price USD',
    'Unit Cost USD', 'Profit per Unit', 'Profit Margin (%)', 'Line Item',
    'Delivery Date', 'Delivery Time', 'Gender', 'Name', 'City',
    'Customer State', 'Continent', 'Birthday', 'Age', 'Brand', 'Color',
    'Subcategory', 'CategoryKey', 'Store State', 'Square Meters',
    'Open Date'
]

# Reorder DataFrame
df = df[column_order]

# Insert into MySQL
try:
    connection = mysql.connector.connect(
        host='localhost',
        user='root',
        password='root',
        database='Datasparkdb'
    )

    if connection.is_connected():
        cursor = connection.cursor()
        print("Connected to Datasparkdb")

        # Create the insert query
        insert_query = f"""
        INSERT INTO merged_global_electronics ({', '.join(f'`{col}`' for col in column_order)})
        VALUES ({', '.join(['%s'] * len(column_order))})
        """

        # Prepare data
        data_to_insert = [tuple(row) for row in df.itertuples(index=False)]

        # Bulk insert
        cursor.executemany(insert_query, data_to_insert)
        connection.commit()
        print(" Data inserted successfully into merged_global_electronics.")

except Error as e:
    print(f" Error: {e}")

finally:
    if connection.is_connected():
        cursor.close()
        connection.close()
        print("MySQL connection closed.")
```
